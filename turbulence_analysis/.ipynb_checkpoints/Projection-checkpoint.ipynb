```python
import numpy as np
import abel
import matplotlib.pyplot as plt
from numba import vectorize
from scipy.special import j0
from scipy.interpolate import interp1d
```

```python
def Beta3D(r):

    beta = 0.7887937465608568
    rc = 3.5583536739429116*142.57359941138424
    out = np.power(1. + (np.abs(r) / rc) ** 2, -3. * beta) 
    
    return out

def BetaProjected(r):
    #Assumed linear gridding for r 
    return abel.hansenlaw.hansenlaw_transform(Beta3D(r), dr=r[1]-r[0], direction='forward')
    
def gen_eta(R500):
    
    r = np.linspace(0,2*R500,10000)
    projected_profile =  interp1d(r,BetaProjected(r), bounds_error=False, fill_value=1., assume_sorted=True)
    
    @np.vectorize
    def volume_profile(r,z):
        
        if np.abs(r)>R500 or np.abs(z)>5*R500: return 0
        else: return Beta3D(np.sqrt(r**2 + z**2))
        
        
        #return Beta3D(np.sqrt(r**2 + z**2))
    
    def eta(r,z):
        
        return volume_profile(r,z)/projected_profile(r)
    
    return eta

eta = gen_eta(1000)

from nicolas.stat_sffov import crhofunc_circap as tf_analytic

beta = 0.7887937465608568
rc = 3.5583536739429116*142.57359941138424

tf_an = tf_analytic('beta', rc, 0, 0, 0, 0, '1000', beta=beta, vectorize=True)
```

```python
def phi(t):
    
    new = np.ones_like(t)
    mask = np.abs(t) <= 3
    t_mask = t[mask]
    new[t<3] = - 1
    new[t>3] = 1
    new[mask] = np.tanh(np.pi/2*np.sinh(t_mask))
    
    return new

def dphi(t):
    
    new = np.ones_like(t)
    mask = np.abs(t) <= 3
    t_mask = t[mask]
    new[t<3] = 0
    new[t>3] = 0
    new[mask] = np.pi/2*np.cosh(t_mask) / np.cosh(np.pi/2 * np.sinh(t_mask))**2
    
    return new
```

```python
r = np.linspace(-10,10,1000)
plt.plot(r, phi(r))
plt.plot(r, dphi(r))
```

```python
Rmax = 1000
Zmax = 1000

def integrand(r,z):
    
    kz = 1e-2
    kr = 1e-2
    
    return (1+r)*eta(Rmax*(1+r)/2, Zmax*(1+z)/2)*np.cos(np.pi*kz*Zmax*(1+z))*j0(np.pi*kr*Rmax*(1+r))*np.pi/2*Rmax**2*Zmax
```

```python
from scipy.special import gamma, kv

beta = 0.7887937465608568
rc = 3.5583536739429116*142.57359941138424
Rfov = 1000
p = Rfov/rc
t = np.linspace(0,p,1000)

def xn_kn(x, n):
    """Computes the function x^n K_n(x).
    """
    
    new = np.ones_like(x)*np.pi*2**(n-1.)/np.sin(n*np.pi)/gamma(1.-n)
    new[x!=0] = x[x!=0]**n * kv(n, x[x!=0])
    
    return new

@np.vectorize
def tf_test(kr, kz):
    
    prefact = 2**(5/2-3*beta)*np.pi*rc**2/gamma(3*beta-1/2)
    
    return prefact*simps(t*j0(2*np.pi*rc*kr*t)*xn_kn(2*np.pi*rc*kz*np.sqrt(1+t**2), 3*beta-1/2),x=t)

kr, kz = np.geomspace(1e-7,1e-2,200), np.geomspace(1e-7,1e-2,200)
kkr, kkz = np.meshgrid(kr, kz)

plt.imshow(tf_test(kkr, kkz)**2, origin='lower')
plt.colorbar()
```

```python
from scipy.special import roots_hermite, roots_legendre, roots_laguerre
import matplotlib.pyplot as plt
from tqdm.notebook import tqdm
from numba import vectorize, njit

order = 60
_x_legendre, _w_legendre = roots_legendre(n=order)
_x_hermite, _w_hermite = roots_hermite(n=order)
_x_laguerre, _w_laguerre = roots_laguerre(n=order)

@njit
def P3D(kr, kz):
    
    return (1+np.sqrt(kr**2 + kz**2)/(1e-4))**(-11/3)

@vectorize(nopython=True)
def shifted_P3D(kr1, kr2, kz):
    
    return np.pi*np.sum(integrand_shifted_P3D(np.pi*(1+_x_legendre), kr1, kr2, kz)*_w_legendre)

@njit
def integrand_shifted_P3D(theta, kr1, kr2, kz):
    
    return (1+np.sqrt(kr1**2 + kr2**2 + kz**2 -2*kr1*kr2*np.cos(theta))/(1e-4))**(-11/3)
```

```python
Peta = tf_test(kkr, kkz)**2

@np.vectorize
def P3D_projected(kr_in):

    return simps(simps(shifted_P3D(kr_in, kkr, kkz)*Peta, x=kz, axis=-1), x=kr, axis=-1)
```
